Q 1: What is Simple Linear Regression (SLR)? Explain its purpose.

-  Simple Linear Regression (SLR) is a statistical and machine learning technique used to model the relationship between one independent variable (X) and one dependent variable (Y) by fitting a straight line to observed data.

   The purpose of SLR is:

     -  To understand how changes in X affect Y

    -   To predict the value of Y for a given value of X

    -   To quantify the strength and direction of the relationship between variables

      SLR assumes that the relationship between X and Y is linear, meaning Y changes at a constant rate with respect to X.

Q2: What are the key assumptions of Simple Linear Regression?

 - For Simple Linear Regression to work correctly, the following assumptions must hold:

 - Linearity:
There must be a linear relationship between the independent variable (X) and dependent variable (Y).

 - Independence of observations:
Each observation should be independent of the others (no autocorrelation).

 - Homoscedasticity:
The variance of errors (residuals) should remain constant across all values of X.

 - Normality of errors:
The residuals should be normally distributed, especially important for hypothesis testing.

 - No significant outliers:
Extreme outliers can disproportionately affect the regression line and reduce model reliability.

Q7: Name and briefly describe three common evaluation metrics for regression models.

 - Mean Absolute Error (MAE):
Measures the average absolute difference between actual and predicted values.

 - Mean Squared Error (MSE):
Measures the average of squared errors, penalizing larger mistakes more heavily.

- R-squared (R¬≤):
Indicates how much variance in the dependent variable is explained by the model.

Q8: What is the purpose of the R-squared metric in regression analysis?

 - R-squared (R¬≤) measures the goodness of fit of a regression model.

 - It represents:

    The proportion of variance in Y explained by X

    Ranges from 0 to 1

 - Interpretation:

    R¬≤ = 0 ‚Üí Model explains nothing

    R¬≤ = 1 ‚Üí Perfect prediction

    Higher R¬≤ ‚Üí Better explanatory power

 - However, R¬≤ does not indicate causation and may be misleading if overfitting exists.



```python
#Q9: Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.
import numpy as np
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 5, 4, 5])

# Create and train model
model = LinearRegression()
model.fit(X, y)

# Print slope and intercept
print("Slope (m):", model.coef_[0])
print("Intercept (b):", model.intercept_)
```

Q10: How do you interpret the coefficients in a simple linear regression model?

 - Slope (m):
For every one-unit increase in X, Y increases or decreases by m units.
Example: If m = 0.6, then Y increases by 0.6 for each unit increase in X.

- Intercept (b):
Represents the expected value of Y when X = 0.
It provides a baseline reference but may not always have real-world meaning.

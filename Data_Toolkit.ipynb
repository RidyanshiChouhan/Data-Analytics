Question 1: What is the difference between multithreading and multiprocessing?

  - Multithreading means running multiple threads within a single process. Threads share the same memory space, which makes context switching faster but can lead to conflicts when accessing shared data.

 - Multiprocessing means running multiple processes simultaneously. Each process has its own memory space, which avoids data corruption but requires more memory and has slower communication between processes.
Use multithreading for I/O-bound tasks (like file or network operations) and multiprocessing for CPU-bound tasks (like heavy computations).








Question 2: What are the challenges associated with memory management in Python?

 - Garbage Collection: Automatic memory cleanup may not immediately free unused objects.

 - Circular References: Objects referencing each other can prevent garbage collection.


 - Memory Leaks: Holding references to unused objects keeps them in memory.


 - Fragmentation: Dynamic allocation can fragment memory, reducing efficiency.


 - Global Interpreter Lock (GIL): Limits true parallelism for threads in Python.

Question 6: What are the differences between NumPy arrays and Python lists?

 - NumPy Array :  Homogeneous (same type), Faster, uses C backend, Contiguous memory, Supports vectorized operations, Fixed size

 - Python List : Heterogeneous (different types), Slower, uses Python objects, Scattered memory allocation, Requires loops, Dynamic size

 - NumPy arrays are better for numerical and scientific computation.

Question 7: Explain the difference between apply() and map() in Pandas.

 - map() → Works on a Series (single column). It applies a function, dictionary, or lambda to each element.
```
df['Age'] = df['Age'].map(lambda x: x + 1)

```
 - apply() → Works on DataFrame or Series. It can apply functions along rows or columns.
```
 df['Total'] = df.apply(lambda row: row['Math'] + row['Science'], axis=1)
```

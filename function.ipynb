1.What is the difference between a function and a method in Python?
  - Function: A block of code that performs a task and is defined using def. It is independent of objects.
  - Method: A function associated with an object/class. It is called using object instance.
  - Example:
    ```
    #Function
    def greet(name):
    return f"Hello, {name}"

    print(greet("Alice"))   # Function call

    # Method
    name = "python"
    print(name.upper())     # Method call
    ```
  - Keywords: Function = independent, Method = tied to object

2.Explain the concept of function arguments and parameters in Python.
  - Parameter: Variable defined inside function signature.
  - Argument: Actual value passed when calling the function.
  - Example:
  ```
def add(a, b):   # a, b = parameters
    return a + b

print(add(3, 5))   # 3, 5 = arguments
```
  - Keywords: parameters = placeholders, arguments = actual values

3.What are the different ways to define and call a function in Python?
  - Normal function: using def.
  - Lambda function: using lambda.
  - Function as object: assign to variable or pass as argument.
  - Example:
```
# Normal
def square(x):
```

5.What are iterators in Python and how do they differ from iterables?
  - Iterable: Object that can be looped over (list, tuple, dict).
  - Iterator: Object that produces values one at a time using __iter__() and __next__().
  - Example:
```
nums = [1,2,3]
it = iter(nums)      # iterator from iterable
print(next(it))      # 1
print(next(it))      # 2
```
  - Keywords: Iterable = can loop, Iterator = generates values step by step

6.Explain the concept of generators in Python and how they are defined.
  - Generators are functions that use yield instead of return.
  - They produce values one at a time (lazy evaluation).
  - Example:
```
def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1

for num in count_up_to(3):
    print(num)  # 1, 2, 3
```


7.What are the advantages of using generators over regular functions?

  - Memory efficient (no need to store all values).
  - Useful for large datasets or streams.
  - Lazy evaluation (generate values when needed).
  - Example:
```
# Generator for even numbers
def even_gen(n):
    for i in range(0, n, 2):
        yield i

for val in even_gen(6):
    print(val)   # 0, 2, 4
```

10.What is the difference between map(), reduce(), and filter()?
   - map(): applies function to each element.
   - filter(): selects elements matching condition.
   - reduce(): combines elements into single value.
   - Example:
```
from functools import reduce

nums = [1,2,3,4]

print(list(map(lambda x: x*2, nums)))    # [2,4,6,8]
print(list(filter(lambda x: x%2==0, nums)))  # [2,4]
print(reduce(lambda a,b: a+b, nums))     # 10
```

````python
#4.Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    # Prime numbers are greater than 1
    if n <= 1:
        return False

    # Check divisibility up to sqrt(n)
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Print all prime numbers from 1 to 200
print("Prime numbers from 1 to 200 are:")
for num in range(1, 201):
    if is_prime(num):
        print(num, end=" ")
````

````python
#5.Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
# Iterator class for Fibonacci sequence
class FibonacciIterator:
    def __init__(self, n_terms):
        self.n_terms = n_terms   # total number of terms required
        self.count = 0           # track how many terms generated
        self.a, self.b = 0, 1    # first two Fibonacci numbers

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.n_terms:
            if self.count == 0:
                self.count += 1
                return self.a
            elif self.count == 1:
                self.count += 1
                return self.b
            else:
                self.a, self.b = self.b, self.a + self.b
                self.count += 1
                return self.b
        else:
            raise StopIteration   # stop when terms are finished

fib = FibonacciIterator(10)   # generate first 10 Fibonacci numbers
for num in fib:
    print(num, end=" ")
````

````python
#10.Create a Python program that uses `filter()` to remove all the vowels from a given string.
text = "Data Analytics with Python"

# Define vowels
vowels = "aeiouAEIOU"

# Use filter to remove vowels
result = "".join(filter(lambda ch: ch not in vowels, text))
print("Without vowels:", result)
````
